```python
import pandas as pd
import numpy as np
from googlemaps import Client
```

```python
data = pd.read_csv('meta/all_sites.csv')
```

```python
data.head()
```

```python
gmaps = Client(key='***')
```

```python
data['address'] = data.apply(lambda row: gmaps.reverse_geocode((row['LAT'],row['LNG'])),axis=1)
```

```python
data['ADDRESS'] = data.apply(lambda row: row['address'][0]['formatted_address'], axis=1)
```

```python
data = data.drop(columns=['address'])
```

```python
data[['STREET','CITY/TOWN/VILLAGE','STATE','COUNTRY']] = data.ADDRESS.str.split(',', expand=True) 
```

```python
data[['STATE','ZIP_CODE']] = data.STATE.str.split(expand=True) 
```

```python
data = data[['SITE_ID', 'INDUSTRY', 'SUB_INDUSTRY', 'SQ_FT', 'LAT', 'LNG', 'TIME_ZONE', 'TZ_OFFSET', 'ADDRESS', 'STREET', 'CITY/TOWN/VILLAGE', 'STATE', 'ZIP_CODE', 'COUNTRY']]
```
